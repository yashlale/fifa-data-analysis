CONVERTING HEIGHT AND WEIGHT TO NUMERIC FORM

```python
df["Height"]=df['Height'].str.strip('cm')
```

```python
df["Height"].unique()
```

```python
def ft_to_cm(x):
    if "'" in x:
        ft=x.replace("'",".").replace('"',"")
        cm=round(float(ft)*30.48,0)
        return int(cm)
    else:
        return int(x)

df["Height"]=df["Height"].apply(ft_to_cm)
df["Height"].unique()    
```

```python
def lbs_to_kg(x):
    if "lbs" in x:
        lbs = x.replace("lbs","")
        kg = round(int(lbs)*0.453,0)
        return int(kg)
    else:
        return int(x)
df["Weight"]=df["Weight"].apply(lbs_to_kg)

```

```python
def money(x):
    if "€" in x:
        x=x.strip("€")
    if "M" in x:
        x=x.strip("M")
        x=float(x)*1000000  
        return int(x) 
    elif "K" in x:
        x=x.strip("K")
        x=float(x)*1000
        return int(x)
    else:
        return int(x)
    
df["Value"]=df["Value"].apply(money)
  
```

CLEANING UP THE CONTRACT COLUMN

```python
df["Contract"].unique()
```

```python
from datetime import datetime, timedelta
```

```python
def type(x):
    if "Free" in x:
        return "Free"
    if "Loan" in x:
        return "Loan"
    if "~" in x:
        return "Contract"
    else:
        return pd.NA
    

def start_time_contract(x):
    if "~" in x:
        return  int(x[:4])
    if "Loan" in x:
        x=x.strip(" On Loan")
        x= datetime.strptime(x, "%b %d, %Y")
        return x.date()
    else:
        return pd.NA

def end_time_contract(type,contract,loan):
    if type == "Contract":
        return int(contract[-4:])
    if type == "Loan":
        loan = datetime.strptime(loan, "%b %d, %Y")
        return loan.date()
    
    else:
        return pd.NA
```

```python
df["Type of contract"] = df["Contract"].apply(type)
```

```python
df["Start year"] = df["Contract"].apply(start_time_contract)
```

```python
df["End year"] = df.apply(lambda row: end_time_contract(row["Type of contract"],row["Contract"],row["Loan Date End"] ), axis=1)
```

```python
df=df[list(df.columns[:7])+list(df.columns[-3:])+list(df.columns[8:-3])]
df=df.drop(columns=["Loan Date End"])
```

```python
df["Attacking"]=round(df["Attacking"]/5,0)
df["Attacking"]=df["Attacking"].astype("int64")
```

```python
df["Skill"]=df["Skill"]/5
df["Skill"]=df["Skill"].astype("int64")
```
